```python
import numpy as np

tags = ["Noun", "Verb"]

words = ["Dogs", "Bark"]

word_to_index = {
    "Dogs": 0,
    "Bark": 1
}

init_prob = np.array([
    0.6,
    0.4
])

transition = np.array([
    [0.7, 0.3],
    [0.4, 0.6]
])

emission = np.array([
    [0.8, 0.2],
    [0.1, 0.9]
])

log_init = np.log(init_prob + 1e-10)
log_trans = np.log(transition + 1e-10)
log_emit = np.log(emission + 1e-10)


def viterbi(sentence):
    T = len(sentence)
    N = len(tags)

    dp = np.full((N, T), -np.inf)
    backpointer = np.zeros((N, T), dtype=int)

    dp[:, 0] = log_init + log_emit[:, sentence[0]]

    for t in range(1, T):
        for current_tag in range(N):
            scores = (
                dp[:, t - 1]
                + log_trans[:, current_tag]
                + log_emit[current_tag, sentence[t]]
            )

            backpointer[current_tag, t] = np.argmax(scores)
            dp[current_tag, t] = np.max(scores)

    best_path = [np.argmax(dp[:, -1])]

    for t in range(T - 1, 0, -1):
        best_path.append(backpointer[best_path[-1], t])

    best_path.reverse()

    return best_path


sentence = ["Dogs", "Bark"]

sentence_index = [word_to_index[word] for word in sentence]

best_tags = viterbi(sentence_index)

print("Sentence:", sentence)
print("\nPredicted Tags:")

for word, tag_index in zip(sentence, best_tags):
    print(word, "->", tags[tag_index])
```

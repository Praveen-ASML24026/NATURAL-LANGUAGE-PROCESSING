```python
s1 = input()
s2 = input()

m = len(s1)
n = len(s2)
arr = []
for i in range(m+1):
    row=[]

    for j in range(n+1):
        row.append(0)

    arr.append(row)

for i in range(m+1):
    arr[i][0] = i

for j in range(n+1):
    arr[0][j] = j

for i in range(1,m+1):
    for j in range(1,n+1):
        if (s1[i - 1] == s2[j - 1]):
            arr[i][j] = arr[i - 1][j - 1]

        else:

            insert = arr[i][j-1]
            delete = arr[j][i-1]
            replace = arr[i-1][j-1]

            minimum = min(insert,delete,replace)

            arr[i][j] = 1+minimum

print("The minimum distance is:  ",arr[m][n])



```
